```python
from openai import OpenAI

client = OpenAI()
```

```python
import time

# Store conversation
messages = [
    {"role": "system", "content": "You are a helpful, friendly chatbot."}
]

def chat():
    print("🤖 ChatGPT Bot (type 'quit' to exit)\n")
    
    while True:
        user_input = input("You: ")
        
        if user_input.lower() in ["quit", "exit", "bye"]:
            print("Bot: Goodbye! 👋")
            break
        
        # Add user message
        messages.append({"role": "user", "content": user_input})
        
        try:
            # API call
            response = client.chat.completions.create(
                model="gpt-4o-mini",
                messages=messages,
                temperature=0.7
            )
            
            reply = response.choices[0].message.content
            
            # Simulate typing effect (optional)
            print("Bot:", end=" ", flush=True)
            for char in reply:
                print(char, end="", flush=True)
                time.sleep(0.01)
            print()
            
            # Save bot response
            messages.append({"role": "assistant", "content": reply})
        
        except Exception as e:
            print("Error:", e)

# Run chatbot
chat()
```

```python
import json

def save_chat(filename="chat_history.json"):
    with open(filename, "w") as f:
        json.dump(messages, f, indent=4)
    print("Chat saved!")

# Call after chat ends
# save_chat()
```
